### 5. Пропуски

Также у нас в исходных данных имеются пропуски в столбце `tenure`. Данный признак практически никак не оказывает влияния на наш целевой признак (корреляция составляет -0.016, что очень мало). Поэтому мы можем заменить пустые значения на сгруппированные по столбцу.

Напишем функцию `na_to_median` для заполнения `NaN`, сгруппированных по одному столбцу с наибольшим показателем корреляции (`is_active_member`)

```python
x = df['tenure'].isna().sum()/df.shape[0]
y = df['tenure'].isna().sum()
print(f'Количество пропущенных значений: {y}, что составляет {x:.2%} от исходных данных')
```

```python
def na_to_median(column1, column2):
    # column1 - столбец группировки
    # column2 - целевой столбец с NA
    
    column1 = df[column1]
    
    for value in column1.unique():
        median = df.loc[column1 == value, column2].median()
        print(f'Для группы {value} медиана равна {median:.1f}')
        df.loc[(df[column2].isna()) & (column1 == value), column2].median()
    
    df[column2] = df[column2].fillna(value=df[column2].median()) # для негруппируемых значений будет выводиться общая медиана
```

### 1. Разделение

Предоставленные данные представляют собой один единый датафрейм, разделения на обучающую, тестовую и валидационную выборки у нас сейчас нет. Поэтому выполним это сейчас с помощью алгоритма `train_test_split` в соотношении 3:1:1.

Введем переменную с признаками и с целевым признаком по всему датафрейму

```python
features = df.drop(['exited'], axis=1)
target = df['exited']
```

Сначала отделим тестовую выборку от тренировочной, которая будет составлять 20% от всех исходных данных, остаток будет обучающая+валидационная. Тестовую будем использовать для оценки эффективности полностью обученного классификатора.

Далее уже разделим на обучающую и валидационную с частью 25%. На обучающей будем выборке будем обучать нашу модель с подгоном параметров. Валидационную будем использовать для настройки параметров классификатора.

```python
features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=.25, random_state=12345)
```

Посмотрим на размеры получившихся выборок:

```python
# добавим все выборки в словарь
arrays = {
    'обучающая' : features_train,
    'валидационная' : features_valid,
    'тестовая' : features_test}
```

```python
# выведем количество содержащихся объектов
for array_name, array in arrays.items():
    print("\033[4m\033[1m\033[31m{}\033[0m содержит объектов:".format(f'{array_name} выборка'), array.shape[0], f'({array.shape[0]/df.shape[0]:.1%})')
```

### 2. Масштабирование

Из вышеописанного обзора данных видно, что помимо категориальных признаков есть и те, которые имеют различный масштаб, что негативно скажется на качестве наших моделей. Стандартизируем их:
* возраст (`age`)
* баланс на счену (`balancе`)

```python
scaler = StandardScaler()
scaler.fit(features_train[numeric]) 

features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

## **III. Построение моделей**

<a id='1st_model'></a>

### 1. Модель «Решающее дерево»

Рассмотрим первый тип модели: используя алгоритм `решающее дерево` и поиграемся с его гиперпараметром глубины для задания максимального количества условий. Через цикл определим лучшую глубину дерева и соответствующую модель данного типа.

```python
best_tree_model_unbalanced = None
best_tree_depth_unbalanced = 0
best_tree_result_unbalanced = 0
tree_model_time_unbalanced = 0

#сначала задал большой разбег с шагом 3-5, чтобы определить промежуток
#for depth in range(1, 60, 5):

#затем более точечно уже
for depth in range(5, 12):
    
# засекаем время
    start_time = time.time()
    
# создаем модельку
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model.fit(features_train, target_train)
    
# кормим ей валидационную выбоку и смотрим на качество опреления целевого признака
    predictions_valid = model.predict(features_valid)
    result = f1_score(target_valid, predictions_valid)
#    print("max_depth =", depth, ": ", end='')
#    print(result, 'F1:', f1_score(target_valid, predictions_valid))
    tree_model_time = (time.time() - start_time)

# сравниваем с лучшим качеством, если находим превосходящую, то перезаписываем модель и отсчитываем время обучения лучшей модели
    if result > best_tree_result_unbalanced:
        best_tree_model_unbalanced = model
        best_tree_result_unbalanced = result
        best_tree_depth_unbalanced = depth
        tree_model_time_unbalanced = tree_model_time

# строим матрицу ошибок лучшей модели
model = DecisionTreeClassifier(random_state=12345, max_depth=best_tree_depth_unbalanced)
model.fit(features_train, target_train)
predictions_valid_tree_model_unbalanced = model.predict(features_valid)
print('\033[4m\033[1m\033[31m{}\033[0m\n'
      .format('Матрица ошибок:'), confusion_matrix(target_valid, predictions_valid_tree_model_unbalanced))

# ROC
probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
# AUC-ROC
auc_roc_tree_model_unbalanced = roc_auc_score(target_valid, probabilities_one_valid)
print('\033[4m\033[1m\033[31m{}\033[0m\n'
      .format('AUC-ROC показатель:'), auc_roc_tree_model_unbalanced)
        
print('\n\033[4m\033[1m\033[31m{}\033[0m'
      .format(f'F1-мера лучшей модели: {best_tree_result_unbalanced} при глубине дерева решений {best_tree_depth_unbalanced}'))

print(f'Время выполнения обучения составляет {tree_model_time_unbalanced:.3f} секунд')
```

Оптимальное **значение глубины**, при котором получаем лучший результат среднее гармоническое полноты и точности в **0.5227** равен **7**.

Согласно матрице ошибок, наша модель достаточно часто ошибается.

### 2. Модель «Случайный лес»

Другим популярным алгоритмом является «Случайный лес», позволяющий построить целый «лес» независимых друг от друга «решающих деревьев», затем выбрать лучшее.

```python
best_forest_model_unbalanced = None
best_forest_result_unbalanced = 0
best_forest_n_estimators_unbalanced = 0
forest_model_time_unbalanced = 0

#сначала задал большой разбег с шагом 3-5-10, чтобы определить промежуток
#for est in range(1, 500, 10):

#затем более точечно уже
for est in range(100, 120, 1):
    
# засекаем время обучения
    start_time = time.time()
    
# создаем модельку
    model = RandomForestClassifier(random_state=12345, n_estimators=est)
    model.fit(features_train, target_train)
    
# кормим ей валидационную выбоку и смотрим на качество опреления целевого признака
    predictions_valid = model.predict(features_valid)
    result = f1_score(target_valid, predictions_valid)
#    print("max_estimators =", est, ": ", end='')
#    print(result)
    forest_model_time = (time.time() - start_time)

# сравниваем с лучшим качеством, если находим превосходящую, то перезаписываем модель и отсчитываем время обучения лучшей модели
    if result > best_forest_result_unbalanced:
        best_forest_model_unbalanced = model
        best_forest_result_unbalanced = result
        best_forest_n_estimators_unbalanced = est
        forest_model_time_unbalanced = forest_model_time

# строим матрицу ошибок лучшей модели
model = RandomForestClassifier(random_state=12345, n_estimators=best_forest_n_estimators_unbalanced)
model.fit(features_train, target_train)
predictions_valid_forest_model_unbalanced = model.predict(features_valid)
print('\033[4m\033[1m\033[31m{}\033[0m\n'
      .format('Матрица ошибок:'), confusion_matrix(target_valid, predictions_valid_forest_model_unbalanced))

# ROC
probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
# AUC-ROC
auc_roc_forest_model_unbalanced = roc_auc_score(target_valid, probabilities_one_valid)
print('\033[4m\033[1m\033[31m{}\033[0m\n'
      .format('AUC-ROC показатель:'), auc_roc_forest_model_unbalanced)

print('\n\033[4m\033[1m\033[31m{}\033[0m'
      .format(f'F1-мера лучшей модели: {best_forest_result_unbalanced} при количестве рассматриваемых деревьев {best_forest_n_estimators_unbalanced}'))

print(f'Время выполнения обучения составляет {forest_model_time_unbalanced:.3f} секунд')
```

Оптимальное **количество деревьев**, при котором получаем лучший результат среднее гармоническое полноты и точности в **0.5526** равно **107**. Код при этом выполняется достаточно долго и обучение происходит медленно.

Согласно матрице ошибок, наша модель достаточно часто ошибается.

### 3. Модель «Логистическая регрессия»

Последним алгоритмом классификатора является «Логистическая регрессия», который сравнивает к какому тарифу ближе объект/клиент (например, основываясь на большей важности возраста нежели наличие кредитной карты) и в зависимости ответа выбирает возможных исход.

```python
best_log_regression_model_unbalanced = None
best_log_regression_result_unbalanced = 0
best_log_regression_iter_unbalanced = 0
log_regression_model_time_unbalanced = 0

for iter in range (1, 100):
    
# засекаем время
    start_time = time.time()

# создаем модельку
    model = LogisticRegression(random_state=12345, solver='lbfgs', max_iter=iter) 
    model.fit(features_train, target_train)
    
# кормим ей валидационную выбоку и смотрим на качество опреления целевого признака
    predictions_valid = model.predict(features_valid)
    result = f1_score(target_valid, predictions_valid)
#    print("max_iter =", iter, ": ", end='')
#    print(result)
    log_regression_model_time = (time.time() - start_time)
    
# сравниваем с лучшим качеством, если находим превосходящую, то перезаписываем модель и отсчитываем время обучения лучшей модели
    if result > best_log_regression_result_unbalanced:
        best_log_regression_model_unbalanced = model
        best_log_regression_result_unbalanced = result
        best_log_regression_iter_unbalanced = iter
        log_regression_model_time_unbalanced = log_regression_model_time

# строим матрицу ошибок лучшей модели
model = LogisticRegression(random_state=12345, solver='lbfgs', max_iter=best_log_regression_iter_unbalanced)
model.fit(features_train, target_train)
predictions_valid_log_regression_model_unbalanced = model.predict(features_valid)
print('\033[4m\033[1m\033[31m{}\033[0m\n'
      .format('Матрица ошибок:'), confusion_matrix(target_valid, predictions_valid_log_regression_model_unbalanced))

# ROC
probabilities_valid = model.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
# AUC-ROC
auc_roc_log_regression_model_unbalanced = roc_auc_score(target_valid, probabilities_one_valid)
print('\033[4m\033[1m\033[31m{}\033[0m\n'
      .format('AUC-ROC показатель:'), auc_roc_log_regression_model_unbalanced)
        
print('\n\033[4m\033[1m\033[31m{}\033[0m'
      .format(f'F1-мера лучшей модели: {best_log_regression_result_unbalanced} при количестве итераций {best_log_regression_iter_unbalanced}'))

print(f'Время выполнения обучения составляет {log_regression_model_time_unbalanced:.3f} секунд')
```

```python
# функция, которая принимает в себя обучающую выборку, целевой признак и количество повторений;
# разделяет обучающую выборку на отрицательные и положительные объекты;
# копируем положительные объекты заданное число раз;
# создаем новую обучающую выборку;
# перемешиваем данные

def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)

    return features_upsampled, target_upsampled
```

```python
# функция, которая принимает в себя обучающую выборку, целевой признак и количество повторений;
# случайным образом отбросывает часть из отрицательных объектов;
# копируем положительные объекты заданное число раз;
# создаем новую обучающую выборку;
# перемешиваем данные

def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
```

```python
fraction = len(target_ones) / len(target_zeros)
fraction
```

```python
features_downsampled, target_downsampled = downsample(features_train, target_train, fraction)
```

Перед тестированием объединим тренировочную и валидационную выборки для небольшого повышения качества.

```python
# создадим пустой список для последующего добавления в обзорную таблицу лучших моделей
f1_tests = []
auc_roc_tests = []


features_mixed = pd.concat([features_train, features_valid], sort=False, axis=0)
target_mixed = pd.concat([target_train, target_valid], sort=False, axis=0)
                
# через цикл, обращаясь к словарю моделей, определим качество моделей по тестовой выборке, дополнительно прогнав модель через объединенную выборку
for name, model in models.items():
    model.fit(features_mixed, target_mixed)
    test_predictions = model.predict(features_test)
    f1_test = f1_score(target_test, test_predictions)
    f1_tests.append(f1_test)
    print(f'Тестовая выборка на модели {name}:', f1_test)
    
    probabilities_test = model.predict_proba(features_test)
    probabilities_one_test = probabilities_test[:, 1]
    auc_roc_test = roc_auc_score(target_test, probabilities_one_test)
    auc_roc_tests.append(auc_roc_test)
    
# создадим новый столбец в выводной таблице по лучшим моделям
models_summary['Качество при тесте'] = f1_tests
models_summary['Сравнение со случайной'] = auc_roc_tests
```
